```python
# filter out categories that are not food-related
restaurants_categories = ["Restaurant", "Bar / Tavern", "Snack Bar", "Special Kitchen", \
                          "Buffet", "Barbeque", "Banquet Kitchen", "Bakery Sales", "Kitchen Bakery", \
                         "Kitchen Bakery", "Portable Bar", "Confection"]
inspection = inspection[inspection['Category Name'].isin(restaurants_categories)]
# filter yelp 
business = business[business['business_id'].isin(inspection['Nearest_yelp_id'])]
```

```python
# write business and inspection data to new csv file 
business.to_csv('final_business.csv', encoding = 'utf-8')
inspection.to_csv('final_inspection.csv', encoding = 'utf-8')
```

```python
# new function to filter a string to only contain letters, also convert the string to be in lower cases
import re
def only_letter(myString):
    result = re.sub(r"[^A-Za-z]+", '', myString)
    return result.lower()
```
