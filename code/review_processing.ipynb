## get the sentiment value for each review

#### use VADER SentimentAnalyzer
have predefined sentiment lexicons and calculate what percentage of the data consists of 1) neutral, 2) positive, 3) negative words. the compound score ranges from -1 to 1 and represents the overall semantic valence of the text

suitable for social media texts because the lexicons contains emojis and slangs

we should only use the compound score then!

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
senti_scores = [[0 for x in range(4)] for i in range(len(reviews))]

# define index for the polarized results
neg = 0
neu = 1
pos = 2
comp = 3
len(reviews)

for i in range(len(reviews)):
    temp_review = reviews.loc[i, 'Content']
    temp = sid.polarity_scores(temp_review)
    senti_scores[i][neg] = temp['neg']
    senti_scores[i][neu] = temp['neu']
    senti_scores[i][pos] = temp['pos']
    senti_scores[i][comp] = temp['compound']
#     reviews.loc[i, 'senti_all'] = senti_scores[i][0:3]
#     reviews.loc[i, 'senti_comp'] = senti_scores[i][3]
# reviews
```

```python
# add senti scores to the dataframe
for i in range(len(reviews)):
    for j in range(3):
        reviews.loc[i, 'senti_all'].append(senti_scores[i][j])
    reviews.loc[i, 'senti_comp'].append(senti_scores[i][3])
```

```python
# use sentiwordnet3.0 on words
# returns a dictionary with .pos, .neg, .obj as keys
def senti_word(s):
    result = {}
    temp = list(swn.senti_synsets(s, 'a'))
    result['neg'] = temp[0].neg_score()
    result['pos'] = temp[0].pos_score()
    result['obj'] = temp[0].obj_score()
    return result
```

```python
# the the first n element of a list if there are more than n items
def take_firstn(newList,n):
    if len(newList) > n:
        return newList[0:n]
    else:
        return newList
```

```python
# use "get_word_forms" package
# returns a single list of possible forms
# input should be SINGLE LIST of keywords share similar meaning
def get_forms(keywords):
    result = []
    # take only the first 3 possible adjetive and verb forms of the word
    for i in range(len(keywords)):
        result.append(keywords[i])
        adjs = take_firstn(list(get_word_forms(keywords[i])['a']), 3)
        verbs = take_firstn(list(get_word_forms(keywords[i])['v']), 3)
        for adj in adjs:
            result.append(adj)
        for verb in verbs:
            result.append(verb)
    # eliminate repetitions
    result = list(set(result))
    return result
```

```python
# add possible adj and verb forms to the keyword list
len_keyword = len(keywords)
temp_keywords = [[''] for i in range(len_keyword)] # keywords with possible endings
for i in range(len_keyword):
    temp_keywords[i] = get_forms(keywords[i])
```

```python
# returns one list of keyword counts for ONE REVIEW
def check_keyword(review):
    result = [0 for x in range(len_keyword)]
    words = nltk.word_tokenize(review)
    tags =  nltk.pos_tag(words)    
    for i in range(len(words)):
        # rule out 'the', 'a', 'is'
        if tags[i] != 'VBZ' and tags[i] != 'DT':
            words[i] = words[i].lower()
            # check spelling
            if d.check(words[i]): # no typos
                for j in range(len_keyword):
                    if words[i] in temp_keywords[j]:
                        result[j] += 1
            else:
                # take the first five possibilities
                possibles = take_firstn(d.suggest(words[i]),5)
                for j in range(len_keyword):
                    commons = set(possibles).intersection(temp_keywords[j])
                    if len(commons) > 0:
                        result[j] += 1
    return result
```

```python
# store the number of occurance in each keyword list in a certain comment
count = [[0 for x in range(len_keyword)] for y in range(len(reviews))] 
for i in range(len(reviews)):
    temp_review = reviews.loc[i, 'Content']
    count[i] = check_keyword(temp_review)
```

```python
# add count results to data frame
# calculate the weighted counts at the same time
for i in range(len(reviews)):
    for j in range(len(count[i])):
        reviews.loc[i, 'counts'].append(count[i][j])
```
